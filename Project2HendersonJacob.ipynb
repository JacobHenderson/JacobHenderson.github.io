```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# url = "https://raw.githubusercontent.com/username/repository/main/fruits_vegetables_prices.csv"
dataFrame = pd.read_csv("Fruit & Vegetable Prices 2020.csv")
# dataFrame.head() # Viewing first few rows of table
```

```python
# Question 1
# What is the distribution of Retail Prices for different forms of fruits and vegetables (Fresh, Canned, Juice, Frozen)?
plt.figure(figsize=(12,6))
sb.boxplot(x='Form', y='RetailPrice', data=dataFrame)
plt.title('Distribution of Retail Prices for different forms')
plt.show()
```

Understanding the differences in prices across different forms of food is helpful for knowing consumer preferences, as well as advising the pricing of products. Here, we can see that dried fruits have notably higher retail prices than other forms. On the opposite end, juice is easily the lowest retail price, making it the cheapest and most attractive to consumers.

```python
# Question 2
# Which Fruit or Vegetable has the highest & lowest average Retail Prices?
avg_prices = dataFrame.groupby('Fruit or Vegetable')['RetailPrice'].mean().sort_values(ascending=False)
avg_prices.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Average Retail Prices for each Fruit or Vegetable')
plt.xlabel('Fruit or Vegetable')
plt.ylabel('Average Retail Price')
plt.show()
```
